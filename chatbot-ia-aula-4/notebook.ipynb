```python
# app/models/schemas.py

from pydantic import BaseModel

# Define o formato de ENTRADA (O que o usuÃ¡rio manda)
class UserMessage(BaseModel):
    message: str

# Define o formato de SAÃDA (O que a API devolve)
class BotResponse(BaseModel):
    response: str
```

```python
# app/routes/chat.py

from fastapi import APIRouter
from app.models.schemas import UserMessage, BotResponse

# Cria um roteador isolado para o Chat
router = APIRouter()

# Define o mÃ©todo POST e o endereÃ§o /chat
@router.post("/chat", response_model=BotResponse)
async def chat(body: UserMessage):
    # Hoje fazemos apenas um ECO (Devolve o que recebeu)
    return {"response": f"Eco: {body.message}"}
```

```python
# app/main.py

from fastapi import FastAPI
from app.routes import chat

# Cria a aplicaÃ§Ã£o
app = FastAPI()

# Conecta as rotas criadas no outro arquivo
# O prefixo organiza a URL: /api/v1/chat
app.include_router(chat.router, prefix="/api/v1")

# Rota de teste simples
@app.get("/")
async def health():
    return {"status": "API Online ğŸš€"}
```
